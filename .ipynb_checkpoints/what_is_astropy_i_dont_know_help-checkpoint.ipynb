```python
tbl1 = ascii.read("astropy_shiz/Young-Objects-Compilation.csv", header_start=1, data_start=2)
tbl1['RA']
```

```python
# fills empty indices with NaN - really useful to know, actually

tbl1['RA'].filled(np.nan)
```

```python
plt.style.use('ggplot')
```

```python
plt.scatter(tbl1["Jmag"] - tbl1["Kmag"], tbl1["Jmag"], alpha=1.0, color='xkcd:barbie pink') # plot J-K vs. J
plt.ylim(reversed(plt.ylim())) # flip the y-axis
plt.xlabel("$J-K_s$", fontsize=20)
plt.ylabel("$J$", fontsize=20);
```

```python
ra = coord.Angle(tbl1['RA'].filled(np.nan)*u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coord.Angle(tbl1['Dec'].filled(np.nan)*u.degree)
```

```python
# it's cool, at any rate

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian);
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian, color='xkcd:pink')
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
#fig.savefig('map.pdf')
ax.grid(True);
```

```python
# oh, what a good idea; I'd never thought of doing that...
# almost some machine learning shit, note use c when doing this instead of color

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian, c=tbl1['Kmag']);
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian, c=tbl1['Kmag'])
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
#fig.savefig('map.pdf')
ax.grid(True);
```
