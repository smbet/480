```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

import astropy.io.fits as fits

import astropy.units as u
import astropy.constants as const

import glob

import os

from photutils.aperture import EllipticalAperture
from photutils.aperture import EllipticalAnnulus
from photutils import aperture_photometry

from matplotlib.colors import LogNorm
```

# First, M101:

```python
plt.style.use('default')
```

```python
M101filepath = glob.glob('./fits/reduced*.fits')
```

```python
M101data_array = []
M101header_array = []

for file in M101filepath:

    hdu = fits.open(file)
    M101data_array.append(hdu[0].data.copy())
    M101header_array.append(hdu[0].header.copy())
    print(os.path.basename(file))

hdu.close()
```

```python
filt_keys = ['halpha', 'g', 'r', 'i', 'z']
```

```python
skyless_ha_M101 = M101data_array[1]
skyless_g_M101 = M101data_array[0]
skyless_r_M101 = M101data_array[3]
skyless_i_M101 = M101data_array[2]
skyless_z_M101 = M101data_array[-1]
```

```python
coord = [2051.134, 2053.268]
```

```python
#ApertureScience = EllipticalAperture(coord, a=850, b=700, theta=(np.pi/4))
```

# Plotting stuff to get a better look at gas content:

```python
fig, ax = plt.subplots(1,3)

fig.set_size_inches(12,6)

fig.tight_layout()

ax[0].axis((1400,3000,1400,3000))
ax[1].axis((1400,3000,1400,3000))
ax[2].axis((1400,3000,1400,3000))

ax[0].imshow(skyless_g_M101, norm=LogNorm(vmin=200), origin='lower', cmap='rainbow')
ax[1].imshow(skyless_ha_M101, norm=LogNorm(vmin=100), origin='lower', cmap='rainbow')
ax[2].imshow(skyless_i_M101, norm=LogNorm(vmin=100), origin='lower', cmap='rainbow')

ax[0].set_title('Sky-Subtracted g Band')
ax[1].set_title('Sky-Subtracted Halpha')
ax[2].set_title('Sky-Subtracted i Band');
```

```python
fig, ax = plt.subplots(1,3)

fig.set_size_inches(12,6)

fig.tight_layout()

ax[0].axis((1400,3000,1400,3000))
ax[1].axis((1400,3000,1400,3000))
ax[2].axis((1400,3000,1400,3000))

ax[0].imshow(skyless_g_M101-skyless_i_M101, norm=LogNorm(vmin=200), origin='lower', cmap='rainbow')
ax[1].imshow(skyless_ha_M101-np.mean(skyless_ha_M101), norm=LogNorm(vmin=100), origin='lower', cmap='rainbow')
ax[2].imshow(skyless_i_M101-skyless_ha_M101, norm=LogNorm(vmin=100), origin='lower', cmap='rainbow')

ax[0].set_title('Just the Blue')
ax[1].set_title('Just the Bright Halpha')
ax[2].set_title('No HII regions!');
```
